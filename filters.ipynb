```python
from skimage import io, exposure, img_as_ubyte
import  matplotlib.pyplot as plt
from skimage.filters import threshold_otsu, gaussian
import numpy as np
```

```python
%matplotlib inline
%cd images
```

```python
img = io.imread("rescaled_equalized.tif")
```

```python
io.imshow(img)
```

```python
zoom = img[2196:2484,2765:2994]
```

```python
io.imshow(zoom)
```

```python
smoothed = gaussian(img, 7)
```

```python
io.imshow(smoothed)
```

```python
zoom_smoothed = smoothed[2196:2484,2765:2994]
```

```python
io.imshow(zoom_smoothed)
```

```python
th = threshold_otsu(smoothed)
binary = np.zeros_like(smoothed,dtype=np.uint8)
binary[smoothed < th] = 0
binary[smoothed >= th] = 255
io.imshow(binary)
```

```python
zoom_binary = binary[2196:2484,2765:2994]
io.imshow(zoom_binary)
```

```python
io.imsave("binary_after_smoothing.tif", binary)
```
